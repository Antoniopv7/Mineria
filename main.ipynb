```python
import pandas as pd
```

```python
df = pd.read_stata('bd-mineria.dta', convert_categoricals=False)
```

```python
df
```

```python
df_cleaned = df[(df['y1'] != -88) & (df['y1'] != 0) & (df['y1'].notnull())]
df_cleaned
```

```python
import numpy as np

# Seleccionar solo las columnas numéricas
df_numerical = df_cleaned.select_dtypes(include=[np.number])

# Ahora, df_numerical contiene solo las columnas numéricas
df_numerical
```

```python
# eliminar las columnas que tengan una correlación mayor a 0.9 con 'y1' (no borrar 'y1')
correlaciones = df_numerical.corr()['y1']
columnas_a_borrar = correlaciones[correlaciones > 0.9].index
columnas_a_borrar = columnas_a_borrar.drop('y1')
df_limpia = df_numerical.drop(columns=columnas_a_borrar)
df_limpia
```

```python
# normalizamos los datos entre 0 a 1.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_limpia_normalized = pd.DataFrame(scaler.fit_transform(df_limpia), columns=df_limpia.columns)
df_limpia_normalized
```

```python
# aplicamos una transformación logarítmica a los datos
import numpy as np
df_normalized_log = np.log1p(df_limpia_normalized)
df_normalized_log
```

```python
# mostrar las columnas que posean al menos un valor NaN
NaN = df_normalized_log.columns[df_normalized_log.isna().any()].tolist()

# contando la longitud de la lista anterior, podemos ver cuántas columnas tienen al menos un valor NaN
len(NaN)
```

```python
# eliminamos las columnas que posean valores NaN
df_normalized_log_clean = df_normalized_log.drop(columns=NaN)
df_normalized_log_clean
```

```python
import pandas as pd
import statsmodels.api as sm

# Definir la función para el método de eliminación hacia atrás
def backward_selection(df, target_var, alpha=0.05):
    variables = df.columns.tolist()
    ordered_pvalues = pd.Series(dtype=float)
    while len(variables) > 0:
        # Añadir una constante al modelo (intercepto)
        vars_with_constant = sm.add_constant(df[variables])
        # Ajustar el modelo OLS
        ols_model = sm.OLS(target_var, vars_with_constant).fit()
        # Obtener los p-valores
        pvals = ols_model.pvalues[1:]  # Excluir la constante
        highest_pval = pvals.max()
        # Eliminar la variable con el p-valor más alto si es mayor que el nivel de significancia
        if highest_pval >= alpha:
            least_significant_var = pvals.idxmax()
            variables.remove(least_significant_var)
        else:
            ordered_pvalues = pvals.sort_values()
            break
    return ordered_pvalues.index.tolist()
```

```python
# Definir la variable objetivo y las variables predictoras
target = df_normalized_log_clean['y1']
data = df_normalized_log_clean.drop(columns='y1')

# Aplicar la función del método backward
Relevant_date = backward_selection(data, target)
```

```python
# cantidad de variables seleccionadas
len(Relevant_date)
```

```python
# Seleccionar las 8 primeras Variables que son relevantes para nosotros
Relevant_date_8 = Relevant_date[:8]
Relevant_date_8
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Dividir los datos en variables predictoras y variable objetivo
X = df_normalized_log_clean[Relevant_date_8]
y = df_normalized_log_clean['y1']

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Definir y ajustar el modelo de regresión lineal
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predecir los valores de la variable objetivo en el conjunto de prueba
y_pred = regressor.predict(X_test)

# Calcular y mostrar el coeficiente de determinación (r cuadrado)
r2 = round(r2_score(y_test, y_pred), 5)
r2
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calcular los residuos
residuos = y_test - y_pred

# Graficar el diagrama de dispersión de residuos
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuos, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')  # Línea horizontal en y=0
plt.title('Diagrama de Dispersión de Residuos')
plt.xlabel('Valores Predichos')
plt.ylabel('Residuos')
plt.grid(True)
plt.show()
```
